## Lue paikallisia netCDF tiedostoja

```python
import arviz as az

# Ladataan ennalta laskettu .nc tiedosto internetistä.
# az.from_netcdf("radon.nc") paikalliselle tiedostolle
idata = az.load_arviz_data("radon")
idata
```

```python
az.style.use("arviz-darkgrid")
az.rcParams["stats.ic_pointwise"] = True
```

## Luo InferenceData PyStan sovituksesta (fit)

```python
# InferenceData on yhteensopiva eri Stan käyttöliittymien kanssa (Kiitos Ari, Allen ja Mitzi!)
import pystan
from scipy import stats
y = stats.norm(10, 1).rvs(1000)
model = """
data {int N; real y[N];} parameters {real mu; real<lower=0> sigma;}
model { mu ~ normal(0, 1); sigma ~ cauchy(0, 1); y ~ normal(mu, sigma);}
"""
stan_model = pystan.StanModel(model_code=model, extra_compile_args=['-flto'])
fit = stan_model.sampling(data={"y":y, 'N':len(y)})
```

```python
idata_pystan = az.from_pystan(
    posterior=fit,
    observed_data=["y"],
)
idata_pystan
```

# ArviZ:n ja InferenceDatan hyödyt
* Johdonmukainen datarakenne tilastollisesta ohjelmointikielestä riippumatta
* Suositun netCDF formaatin etuina tallennettavuus
* Tulosten jakaminen rohkaisee toistettavuuteen
* Tilastollisten ohjelmointikielien kehittäjät voivat keskittyä kieleen ja sen ominaisuuksiin, jolloin otannan jälkeisen diagnostiikan ja kuvaajien luomiseen tarvittavaa koodia ei tarvitse tuottaa

# InferenceDatan rajoitukset
* InferenceData on suunniteltu MCMC otannalle
* Ei tee bayesialaisesta työnkulusta vielä täysin toistettavaa
